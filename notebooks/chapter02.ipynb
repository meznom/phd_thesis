```python
def dos(Es_, Emin=0, Emax=10, offset=0, deltaE=1E-4, Eoffset=None,gamma=None):
    Es = Es_ - Es_[offset]
    if Eoffset is not None:
        Es += Eoffset
    # construct a matrix with zero everywhere in the range Emin to Emax, 
    # and only a few delta peaks where the energy eigenvalues are located
    xs = np.arange(Emin,Emax,deltaE)
    m = np.row_stack((
        np.column_stack((xs,np.zeros(xs.shape))),
        np.column_stack((Es,np.ones(Es.shape)))
    ))
    m = m[m[:,0].argsort()]
    # simply convolve with a Lorentzian: this gives us the density of states
    # the resolution deltaE must be much better than the peak width gamma
    lorentz = lambda x,x0,gamma: gamma*gamma / ((x-x0)*(x-x0) + gamma*gamma)
    x0 = Emin+(Emax-Emin)/2.0
    if not gamma:
        gamma = (Emax-Emin)/1000.0
    convolved = np.convolve(m[:,1], lorentz(m[:,0],x0,gamma),'same')
    m = np.column_stack((m[:,0],convolved))
    m = m[m[:,1] != 0]
    return m

def dosplot(p, dos1, dos2, label1, label2, xlim=None, ylim=None):
    p.plot(dos1[:,0],dos1[:,1],label=label1)
    p.plot(dos2[:,0],-dos2[:,1],label=label2)
    p.set_xlabel('energy')
    p.set_ylabel('state count')
    if xlim: p.set_xlim(xlim)
    if ylim:
        p.set_ylim(ylim)
        p.set_yticks(range(ylim[0],ylim[1]+1))
        p.set_yticklabels([abs(i) for i in range(ylim[0],ylim[1]+1)])
```

```python
T = 0.02
Emin,Emax,deltaE = (-1,10,1E-3)
E_f = rs['mu',0]['model','QcaFixedCharge'][0].table[0,0]
E_b = rs['mu',0]['model','QcaBond'][0].table[0,0]
dos_f = dos(E_f,Emin,Emax,deltaE=deltaE,gamma=0.5*T)
dos_b = dos(E_b,Emin,Emax,deltaE=deltaE,Eoffset=E_f[1]-E_f[0],gamma=0.5*T)

fig,p = plt.subplots(1,1)
dosplot(p, dos_f,dos_b,'Fixed','Bond',(-1,10),(-1,3))
p.legend()
```

```python
T = 0.02
Emin,Emax,deltaE = (-1,50,1E-3)
E_f = rs['mu',250]['model','QcaFixedCharge'][0].table[0,0]
E_g = rs['mu',250]['model','QcaGrandCanonical'][0].table[0,0]
dos_f = dos(E_f,Emin,Emax,deltaE=deltaE,gamma=0.5*T)
dos_g = dos(E_g,Emin,Emax,deltaE=deltaE,gamma=0.5*T)

fig,p = plt.subplots(1,1)
dosplot(p, dos_g,dos_f,'Grand','Fixed',(-1,50),(-20,30))
p.legend()
```

```python
e = coma.Experiment('../experiments/experiment.000042/')
rs = e.retrieve_results(
    (
        ('T', 'parameters/T'),
        ('P', 'results/P'),
        ('N', 'results/N')
    ),
    (
        ('model', 'parameters/model'),
        ('N', 'parameters/layout/N'),
        ('V1', 'parameters/layout/V1'),
        ('V0', 'parameters/V0')
    ))
rs
```

```python
ls = {'QcaBond': 'Bond','QcaFixedCharge': 'Fixed','QcaGrandCanonical': 'Grand'}

def PTplot(p,rs):
    for r in rs:
        d = r.table
        p.plot(d[:,0], d[:,2], label=ls[r.parameters.model])
    p.set_xlabel('$T$')
    p.set_ylabel('polarization $P_2$')
    p.legend(loc='upper right')

def NTplot(p,rs):
    d = rs['model','QcaGrandCanonical'][0].table
    p.plot(d[:,0], d[:,7],label='Grand. Cell 1',color='blue')
    p.plot(d[:,0], d[:,12],label='Grand. Cell 2',color='red')
    d = rs['model','QcaFixedCharge'][0].table
    p.plot(d[:,0], d[:,7],label='Fixed. Cell 1,2',color='green')
    p.set_xlabel('$T$')
    p.set_ylabel('particle number $N$')
    p.legend(loc='upper right')
```

```python
rs.sort(key=lambda i: i.parameters.model, reverse=True)
fig,p = plt.subplots(1,1)
PTplot(p, rs['N',2]['V1',20]['V0',1E6])
```
